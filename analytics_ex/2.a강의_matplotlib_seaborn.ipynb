## 문제 2. 엔진 크기와 이산화탄소 배출량의 관계를 분석하세요.

- (히트맵으로) displ 이 엔진사이즈이고, co2 가 이산화탄소 배출량입니다. 트랜드를 분석하세요

```python
# 데이터 불러오기 (예: mpg.csv)
df = pd.read_csv('datas/fuel_econ.csv')

# displ(엔진크기), co2(이산화탄소배출량) 간의 상관계수 계산
corr = df[['displ', 'co2']].corr()

# 히트맵 시각화
plt.figure(figsize=(5, 4))
sb.heatmap(corr, annot=True, cmap='RdBu_r', vmin=0.8, vmax=1.0)
plt.title('엔진 크기와 이산화탄소 배출량의 관계')
plt.show()
```

## 문제3 displ 배출량의 이상치 확인

- seabor의 박스플랏
- 'displ' 컬럼의 데이터에 이상치가 있는지 확인하기

```python
# 데이터 불러오기 (예시)
df = pd.read_csv('datas/fuel_econ.csv')

# 박스플롯으로 이상치 확인
plt.figure(figsize=(6,4))
sb.boxplot(x=df['displ'])
plt.title('displ 배출량의 이상치 확인')
plt.show()

# 통계적으로 이상치 값 직접 확인 (선택)
Q1 = df['displ'].quantile(0.25)
Q3 = df['displ'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['displ'] < Q1 - 1.5 * IQR) | (df['displ'] > Q3 + 1.5 * IQR)]

print("이상치 개수:", len(outliers))
print(outliers['displ'])
```
